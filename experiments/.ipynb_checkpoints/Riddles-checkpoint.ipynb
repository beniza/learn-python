# Factorial


````python
def fact(n):
  if(n <= 1):
    return 1
  return n * fact(n-1)
````

````python
fact(4)
````

## Find the next fibenocci number

The relationship between two consequent fibenocci numbers $n1$ and $n2$:

$$\frac{n1}{n2} = 1 + \frac{\sqrt{5}}{2} = 1.618...$$

```For smaller numbers, the result will be slightly off. However as the numbers gets bigger, the closer they get to this answer``` 

````python
from math import *
````

````python
def next_fib(n):
    return round(n * ((1 + sqrt(5))/2.0)) # The magic equation
````

````python
n = int(input('Enter the num: '))
print(next_fib(n))
````

## Riddle


````python
for i in range(1,5):
    for j in range(10):
        if (2 * (i * 10 + j) == (j * 10 + (2 * i))):
            print(i,j,j,2*i)
````

### Oneliner using list comprehension
The above snippet can be expressed using list comprehension as shown in the code block below.

````python
[(x, y, y, 2*x) for x in range(1,5) for y in range(10) if (2 * (x * 10 + y) == (y * 10 + (2 * x)))][0]
````

ചോദ്യം: അയാളുടെ കയ്യിലെ പൂക്കളുടെ എണ്ണമെത്ര?   👮

````python
x = 1
result = False
while(not result):
    for s in range(x + 1, x * 2): 
        # x + 1 becasue, the lower bound must be more than the total number of flowers he's starting with, otherwise the answer won't converge
        # x * 2 because, the upper bound cant' be more that 2 times the number of flowers as that's the maximum he can present at the temple.
        if (8 * x - 7 * s == 0): # this equation is obtained by solving the equation (2 * ((2 * ((2 * x) - s)) - s)) - s
            print("Start with {} flowers and present {} flower at each temple".format(x, s))
#             result = True # Comment this line if you want more that one results
    x += 1
````

**What's happening here**

Step 1: express the problem in a mathematical notation, so that you can solve it using a `code snippet`. The devotee has to start his pilgrimage with an unknown set of flowers. As he take dip in each of the ponds, the number flowers he has will get doubled. So let's assume he starts his pilgrimage with $x$ number of flower.

He then submits $s$ number of flowers at each of the temple. Please remember that $s$ is same, regardless of the number of flowers he has.

We also know that before entering into the temple he has to take dip at the pond, so at the first temple itself he has `two times` flowers. In other words, he has $2x$ flowers in his hand. He then submits $s$ number of flowers at the temple, and he repeats the ritual (dip + submission of flowers) 2 more times. After visiting the first temple he has $2x - s$ number of flowers in his hand. So the final equation is that 

$$(2 ((2 ((2x) - s)) - s)) - s = 0$$

$$\implies \left[ 2 \times [2 \times [(2x -s)] - s] -s = 0\right] $$

$$\implies (8 \times x) - (7 \times s) = 0 $$

$$\implies (8 \times x) = (7 \times s)$$ <div style="text-align: right"> We have used this eqn. in the code $8x - 7s = 0$</div>

$$ \therefore \frac{8}{7} = \frac{s}{x} $$

$$\implies x = 7 \space \& \space  s = 8$$
